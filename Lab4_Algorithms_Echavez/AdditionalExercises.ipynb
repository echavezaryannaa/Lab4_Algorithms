```python
# Algorithm Complexity Profiler

import time

student_id = "TUPM-25-0201"
target = int(student_id[-2:])

main_list = list(range(1, 201))
duplicate_list = [3, 7, 2, 9, 5, 7]

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

def find_duplicate(arr):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] == arr[j]:
                return arr[i]
    return None

start = time.time()
linear_result = linear_search(main_list, target)
linear_time = time.time() - start

start = time.time()
binary_result = binary_search(main_list, target)
binary_time = time.time() - start

start = time.time()
duplicate_result = find_duplicate(duplicate_list)
quadratic_time = time.time() - start

print("\n=== Algorithm Complexity Profiler ===")
print(f"Target Value: {target}\n")

print("{:<20} {:<15} {:<15}".format("Algorithm", "Result", "Time (seconds)"))
print("-" * 50)

print("{:<20} {:<15} {:<15.10f}".format("Linear Search", linear_result, linear_time))
print("{:<20} {:<15} {:<15.10f}".format("Binary Search", binary_result, binary_time))
print("{:<20} {:<15} {:<15.10f}".format("Quadratic (Duplicate)", duplicate_result, quadratic_time))

print("\n=== Analysis ===")
print("Linear Search checks each element one by one → O(n)")
print("Binary Search splits the list repeatedly → O(log n) → faster")
print("Quadratic algorithm uses nested loops → O(n^2) → slowest")

if binary_time < linear_time:
    print("\nBinary Search is faster than Linear Search for large datasets.")
```

```python
# Recursive vs. Iterative Strategy Analysis

import time

student_id = "TUPM-25-0201"
n = int(student_id[-1]) + 15

def fibonacci_iterative(n):
    if n <= 1:
        return n
    
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    
    return b

start = time.time()
result = fibonacci_iterative(n)
execution_time = time.time() - start

print("\n=== Recursive vs Iterative Strategy Analysis ===\n")

print("{:<15} {:<20} {:<20}".format("Problem Size (n)", "Fibonacci Result", "Time (seconds)"))
print("-" * 60)

print("{:<15} {:<20} {:<20.10f}".format(n, result, execution_time))

print("\n=== Analysis ===")
print("The iterative Fibonacci algorithm follows a linear time complexity O(n).")
print("This means the execution time increases proportionally as n increases.")
print("Each step computes the next Fibonacci number using previous values,")
print("making it efficient compared to recursive approaches.")
```

```python
# Sorting Algorithm Visual Comparator

import time
import random

student_id = "TUPM-25-0201"
size = int(student_id[-3:]) + 50

original_list = random.sample(range(1, size * 5), size)

def bubble_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    
    return comparisons, swaps

def selection_sort(arr):
    comparisons = 0
    swaps = 0
    n = len(arr)
    
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    
    return comparisons, swaps

def insertion_sort(arr):
    comparisons = 0
    swaps = 0
    
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        
        while j >= 0:
            comparisons += 1
            if arr[j] > key:
                arr[j + 1] = arr[j]
                swaps += 1
                j -= 1
            else:
                break
        
        arr[j + 1] = key
    
    return comparisons, swaps

arr1 = original_list.copy()
start = time.time()
b_comp, b_swap = bubble_sort(arr1)
b_time = time.time() - start

arr2 = original_list.copy()
start = time.time()
s_comp, s_swap = selection_sort(arr2)
s_time = time.time() - start

arr3 = original_list.copy()
start = time.time()
i_comp, i_swap = insertion_sort(arr3)
i_time = time.time() - start

print("\n=== Sorting Algorithm Comparator ===\n")
print(f"List Size: {size}\n")

print("{:<20} {:<15} {:<15} {:<15}".format("Algorithm", "Comparisons", "Swaps", "Time (s)"))
print("-" * 70)

print("{:<20} {:<15} {:<15} {:<15.6f}".format("Bubble Sort", b_comp, b_swap, b_time))
print("{:<20} {:<15} {:<15} {:<15.6f}".format("Selection Sort", s_comp, s_swap, s_time))
print("{:<20} {:<15} {:<15} {:<15.6f}".format("Insertion Sort", i_comp, i_swap, i_time))

print("\n=== Recommendation ===")

times = {
    "Bubble Sort": b_time,
    "Selection Sort": s_time,
    "Insertion Sort": i_time
}

best = min(times, key=times.get)

print(f"The most efficient algorithm based on execution time is: {best}")
print("Insertion Sort often performs better for partially sorted data,")
print("while Bubble and Selection Sort are generally less efficient (O(n^2)).")
```
