```python
# Step 1: Define an algorithm in pseudocode as comments
# Algorithm to find the largest number in a list
# 1. Assume the first number is the largest
# 2. Compare each number with the current largest
# 3. If a number is larger, update the largest
# 4. Return the largest number
# Step 2: Implement it in Python
numbers = [10, 5, 20, 8, 15]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
        
print("Largest number:", largest)
```

```python
# Step 1: Define a simple linear loop
def print_items(items):
    for item in items:
        print(item)
# Step 2: Observe the number of operations relative to input size
sample = list(range(100))
print_items(sample) # O(n)
```

```python
# Try It no.2:
# Function 1: Nested loop - O(n²)
def print_pairs(items):
    for i in items:           # Outer loop: n times
        for j in items:       # Inner loop: n times
            print(i, j)       # Total operations ~ n * n = n²

# Function 2: Print first item - O(1)
def print_first_item(items):
    if items:                 # Single operation, independent of n
        print(items[0])

# Example usage
sample = [1, 2, 3]

print("Pairs (O(n²)):")
print_pairs(sample)

print("\nFirst item (O(1)):")
print_first_item(sample)
```

```python
# Step 1: Implement linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
# Step 2: Test the function
data = [3, 5, 2, 4, 9]
index = linear_search(data, 4)
print("Found at index:", index)
```

```python
# Try It no.3:
# Linear search with comparison count
def linear_search_count(arr, target):
    comparisons = 0  # Initialize counter
    for i in range(len(arr)):
        comparisons += 1          # Count this comparison
        if arr[i] == target:
            return i, comparisons  # Return index and count
    return -1, comparisons          # Target not found

# Test the function
data = [3, 5, 2, 4, 9]
target = 4

index, comp = linear_search_count(data, target)
print(f"Found at index: {index} after {comp} comparisons")
```

```python
# Step 1: Implement binary search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
# Step 2: Test with a sorted list
sorted_data = [1, 3, 5, 7, 9, 11]
result = binary_search(sorted_data, 7)
print("Found at index:", result)
```

```python
# Step 1: Implement bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
# Step 2: Test sorting
to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
print("Sorted list:", to_sort)
```

```python
#Try It no.5:
# Bubble sort with swap counter
def bubble_sort_count(arr):
    n = len(arr)
    swap_count = 0  # Initialize swap counter
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap
                swap_count += 1  # Count this swap
                
    return swap_count

# Test the function
to_sort = [64, 34, 25, 12, 22, 11, 90]
swaps = bubble_sort_count(to_sort)
print("Sorted list:", to_sort)
print("Total swaps:", swaps)
```

```python
# Step 1: Implement selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
# Step 2: Test and observe performance
sample_data = [29, 10, 14, 37, 13]
selection_sort(sample_data)
print("Sorted list:", sample_data)
```

```python
# Try It no.6:
# Bubble Sort with comparison and swap counters
def bubble_sort_metrics(arr):
    n = len(arr)
    comparisons = 0
    swaps = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    return comparisons, swaps

# Selection Sort with comparison and swap counters
def selection_sort_metrics(arr):
    n = len(arr)
    comparisons = 0
    swaps = 0
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
            swaps += 1
    return comparisons, swaps

# Test data
data1 = [29, 10, 14, 37, 13]
data2 = data1.copy()

# Run metrics
bubble_comp, bubble_swaps = bubble_sort_metrics(data1)
selection_comp, selection_swaps = selection_sort_metrics(data2)

print("Bubble Sort -> Comparisons:", bubble_comp, ", Swaps:", bubble_swaps)
print("Selection Sort -> Comparisons:", selection_comp, ", Swaps:", selection_swaps)
```

```python
# Step 1: Implement insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
# Step 2: Test and measure performance
arr = [12, 11, 13, 5, 6]
insertion_sort(arr)
print("Sorted array:", arr)
```

```python
# Try It No.7:
def insertion_sort_metrics(arr):
    comparisons = 0
    shifts = 0
    n = len(arr)
    
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        
        # Compare key with sorted portion
        while j >= 0:
            comparisons += 1  # Every comparison in while
            if key < arr[j]:
                arr[j + 1] = arr[j]  # Shift element
                shifts += 1
                j -= 1
            else:
                break
        arr[j + 1] = key
    
    return comparisons, shifts

# Test with an already sorted list
sorted_list = [5, 6, 11, 12, 13]
comparisons, shifts = insertion_sort_metrics(sorted_list.copy())
print("Sorted list remains:", sorted_list)
print("Comparisons:", comparisons, "Shifts:", shifts)
```
